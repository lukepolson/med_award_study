```python
def get_p(lwr_yr=-np.inf, upr_yr=np.inf, df=df):
    cond = (df['Year']<upr_yr) & (df['Year']>lwr_yr)
    fem = df[cond][['Male', 'Female']].sum()[1]
    male = df[cond][['Male', 'Female']].sum()[0]
    n = fem+male
    p = fem/n
    p_err = np.sqrt(p*(1-p)/n)
    return np.array([p, p_err])
```

```python
yrs = [-np.inf, 1990, 2000, 2010, np.inf]
p_arr = np.array([get_p(yrs[i], yrs[i+1]) for i in range(len(yrs)-1)])
```

```python
p = p_arr.T[0]
p_err = p_arr.T[1]
yrs_l = ['<1990', '1990-2000', '2000-2010', '>2010']
```

```python
plt.figure(figsize=(8,4))
plt.errorbar(yrs_l, 100*p, 100*p_err, fmt='go--')
plt.ylabel('% Females With Awards')
plt.xlabel('Year')
plt.grid()
```
