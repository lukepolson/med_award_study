```python
import numpy as np
import pandas as pd
from ethnicolr import census_ln, pred_census_ln, pred_fl_reg_name, pred_wiki_name
import matplotlib.pyplot as plt
import matplotlib
plt.style.use(['science', 'notebook'])
```

```python
df = pd.read_csv('sample_names.csv', names=['name'], skiprows=1)
arr = df.name.to_numpy()
```

```python
arr = np.array([a.split(',')[0] for a in arr])
```

```python
arr = np.array([a.split(':')[-1] for a in arr])
```

```python
def remove_char(arr, char):
    return ' '.join(list(filter(lambda x:char not in x, arr.split())))
```

```python
arr = np.vectorize(remove_char)(arr, '*')
arr = np.vectorize(remove_char)(arr, '.')
arr = np.vectorize(remove_char)(arr, '(')
```

```python
f_name = [a.split(' ')[0] for a in arr]
l_name = [a.split(' ')[-1] for a in arr]
df = pd.DataFrame({'fname':f_name, 'lname':l_name})
```

```python
raw_data_full = pred_fl_reg_name(df, 'lname', 'fname')
raw_data_full.replace('(S)', 0, inplace=True)
```

```python
raw_data_full['race'].value_counts()
```

# Some Statistics

```python
df = pd.read_csv('master.csv', delimiter=',')
df['Year'] = pd.to_numeric(df['Year'], errors="coerce")
df['Gender'] = pd.to_numeric(df['Gender'], errors="coerce")
```

```python
df['Male'] = df['Gender']==1
df['Female'] = df['Gender']==2
```

```python
years = df[['Year']].to_numpy().ravel().astype(int)
```

```python
plt.figure(figsize=(8,3))
plt.hist(years[years>0], bins=30)
plt.xlabel('Years')
plt.ylabel('# of Awards')
plt.show()
```

```python
ax = df['Society'].value_counts().plot.bar(fontsize=8)
ax.set_ylabel('# Awards')
```

```python
df_sum = df.groupby('Society').sum()[['Male', 'Female']]
df_sum['ratio'] = df_sum['Female']/df_sum['Male']
df_sum['p'] = df_sum['Female']/(df_sum['Male']+df_sum['Female'])
df_sum['p_err'] = np.sqrt(df_sum['p']*(1-df_sum['p'])/(df_sum['Male']+df_sum['Female']))
```

```python
df_sum
```

```python
def get_p(lwr_yr=-np.inf, upr_yr=np.inf, df=df):
    cond = (df['Year']<upr_yr) & (df['Year']>lwr_yr)
    fem = df[cond][['Male', 'Female']].sum()[1]
    male = df[cond][['Male', 'Female']].sum()[0]
    n = fem+male
    p = fem/n
    p_err = np.sqrt(p*(1-p)/n)
    return np.array([p, p_err])
```

```python
yrs = [-np.inf, 1990, 2000, 2010, np.inf]
```

```python
p_arr = np.array([get_p(yrs[i], yrs[i+1]) for i in range(len(yrs)-1)])
```

```python
p = p_arr.T[0]
p_err = p_arr.T[1]
yrs_l = ['<1990', '1990-2000', '2000-2010', '>2010']
```

```python
plt.figure(figsize=(8,4))
plt.errorbar(yrs_l, 100*p, 100*p_err, fmt='go--')
plt.ylabel('% Females With Awards')
plt.xlabel('Year')
plt.grid()
```

```python
df['Society']=='Society of Nuclear Medicine & Molecular Imaging'
```

```python
yrs = [-np.inf, 1990, 2000, 2010, np.inf]
soc1 = 'Society of Nuclear Medicine & Molecular Imaging'
soc2 = 'Society for Pediatric Radiology'
p1, p_err1 = np.array([get_p(yrs[i], yrs[i+1], df = df[df['Society']==soc1]) for i in range(len(yrs)-1)]).T
p2, p_err2 = np.array([get_p(yrs[i], yrs[i+1], df = df[df['Society']==soc2]) for i in range(len(yrs)-1)]).T
```

```python
plt.figure(figsize=(8,4))
plt.errorbar(yrs_l, 100*p1, 100*p_err1, fmt='go--', label='Nuc Med')
plt.errorbar(yrs_l, 100*p2, 100*p_err2, fmt='ro--', label='Ped Rad')
plt.ylabel('% Females With Awards')
plt.xlabel('Year')
plt.grid()
plt.legend()
```
